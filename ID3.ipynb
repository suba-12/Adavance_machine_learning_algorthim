```python
from chefboost import Chefboost as chef
```

```python
import pandas as pd
```

```python
df = pd.read_csv("D:/LAB/AML/balloondataset.csv")
```

```python
df
```

```python
config = {'algorithm' :"ID3"}
model = chef.fit(df,config = config,target_label = 'Inflated')
```

```python
df
```

```python
prediction = chef.predict(model,param =['PURPLE','LARGE','DIP','CHILD'])
```
